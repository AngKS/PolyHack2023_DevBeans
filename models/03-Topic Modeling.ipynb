## Zero Shot Topic Modeling (Predefined Topics)

```python
# !pip install sentence-transformers
```

```python
import pandas as pd
pd.set_option('display.max_colwidth', None)

from transformers import pipeline
```

```python
classifier = pipeline(task="zero-shot-classification", model="facebook/bart-large-mnli", device=0)
```

```python
text = ("Apple Inc. decided to hire John Smith as their new CEO. "
        "The event will take place in San Francisco on 2022-07-23.")
```

```python
import spacy

# Load English tokenizer, tagger, parser, NER and word vectors
nlp = spacy.load("en_core_web_sm")

def ner(text):
    doc = nlp(text)

    # Find named entities, phrases and concepts
    entities = []
    for entity in doc.ents:
        entities.append([entity.text, entity.label_])

    result = {"entities": entities}
    return result

entities = ner(text)
print(entities)
```

```python
labels = list(map(lambda x: x[0], entities['entities']))
print(labels)
```

# Single Topic

```python
hypothesis_template = "The topic of this post is about {}."

topic_prediction = classifier([text], labels, hypothesis_template=hypothesis_template)
topic_prediction = pd.DataFrame(topic_prediction)
topic_prediction.head()
```

```python
topic_prediction['predicted_topic'] = topic_prediction['labels'].apply(lambda x: x[0])
topic_prediction['predicted_topic_score'] = topic_prediction['scores'].apply(lambda x: x[0])
topic_prediction.head()
```

# Multi Topic

```python
hypothesis_template = "The topic of this post is about {}."

topic_prediction = classifier([text], labels, hypothesis_template=hypothesis_template, multi_label=True)
topic_prediction = pd.DataFrame(topic_prediction)
topic_prediction.head()
```

```python
threshold = 0.9

topic_prediction = topic_prediction.set_index('sequence').apply(pd.Series.explode).reset_index()
topic_prediction = topic_prediction[topic_prediction['scores'] >= threshold]
topic_prediction.head()
```
